# Combining Samples

```python
import pandas as pd
df1=pd.read_csv('./Data/Acir_mT19_allSamples_geneAbundance_tpm.txt',sep='\t').set_index('geneID')
modified_list = [item[:-2] for item in list(df1.columns)]
df1.columns=modified_list
result_df = df1.groupby(level=0, axis=1).sum()
result_df1=result_df.reset_index().set_index('geneID')
result_df1
```

```python
df2=pd.read_csv('./Data/Mar_mT19_allSamples_geneAbundance_tpm.txt',sep='\t').set_index('geneID')
modified_list = [item[:-2] for item in list(df2.columns)]
df2.columns=modified_list
result_df = df2.groupby(level=0, axis=1).sum()
result_df2=result_df.reset_index().set_index('geneID')
result_df2
```

# Geneid modification

```python
import pandas as pd
Taihu98_gene_expression=pd.read_csv('./Data/Taihu98_gene_expression.csv').set_index('Unnamed: 0').rename_axis(None)
Taihu98_gene_expression.index=Taihu98_gene_expression.index.str.split('_').str[:2].str.join('_')
Taihu98_gene_expression#.to_csv('Taihu98_gene_expression_geneid_modified.csv')
Taihu98_snp=pd.read_csv('./Data/Taihu98_snp.csv').set_index('Unnamed: 0').rename_axis(None)
Taihu98_snp.index=Taihu98_snp.index.str.split('_').str[:2].str.join('_')
Taihu98_snp#.to_csv('Taihu98_snp_geneid_modified.csv')
```

```python
Acir_gene_expression=pd.read_csv('./Data/Acir_gene_expression.csv').set_index('Unnamed: 0').rename_axis(None)
Acir_gene_expression.index=Acir_gene_expression.index.str.split('_').str[:2].str.join('_')
Acir_gene_expression#.to_csv('Acir_gene_expression_geneid_modified.csv')
Acir_snp=pd.read_csv('./Data/Acir_snp.csv').set_index('Unnamed: 0').rename_axis(None)
Acir_snp=Acir_snp.apply(pd.to_numeric, errors='coerce').fillna(0).astype('int64')
Acir_snp.index=Acir_snp.index.str.split('_').str[:2].str.join('_')
Acir_snp#.to_csv('Acir_snp_geneid_modified.csv')
```
