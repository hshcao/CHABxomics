# Taihu98

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
Taihu98_gene_expression=pd.read_csv('./Processed_Data/Taihu98_gene_expression_geneid_modified.csv').set_index('Unnamed: 0').rename_axis(None)
Taihu98_snp=pd.read_csv('./Processed_Data/Taihu98_snp_geneid_modified.csv').set_index('Unnamed: 0').rename_axis(None)
print(Taihu98_snp.dtypes,Taihu98_gene_expression.dtypes)
Taihu98_gene_expression;Taihu98_snp
```

```python
merged_Taihu98=pd.merge(Taihu98_snp,Taihu98_gene_expression,left_on=Taihu98_snp.index,right_on=Taihu98_gene_expression.index)
merged_Taihu98=merged_Taihu98.set_index('key_0').rename_axis(None)
merged_Taihu98
```

```python
samples=[]
for i in range(18):
    snp_id=merged_Taihu98.columns[0]
    gene_expression_id='mT19_'+snp_id.split('G')[1]
    sample=merged_Taihu98[[snp_id,gene_expression_id]]
    sample=sample.sort_values(by=snp_id,ascending=True).reset_index().drop('index',axis=1)
    sample.columns=['snp','gene_expression']
    samples.append(sample)
samples[0]
```

```python
df=pd.concat(samples,join='outer',ignore_index=True,axis=0)
df=df.sort_values(by='snp')
df=df.reset_index().drop('index',axis=1)
df
```

```python
####################################################################################
def plot_boxplot_grouped(data_list, labels, log_scale=True, y_upper_limit=None, snp_ranges=None, title=None):
    """
    Box Plot with Grouped SNP Ranges

    Parameters:
    - data_list: List containing data for each boxplot
    - labels: Labels for each boxplot
    - log_scale: Whether to apply logarithmic transformation to the data, default is True
    - y_upper_limit: Upper limit for the y-axis, default is None
    - snp_ranges: List of SNP value ranges, e.g., [(1, 3), (4, 6), (7, 7)], default is None
    - title: Title for the plot, default is None
    """
    # Log transformation
    if log_scale:
        data_list = [np.log1p(data) for data in data_list]

    # Group data based on SNP ranges
    grouped_data = []
    grouped_labels = []
    for rng in snp_ranges:
        start, end = rng
        grouped_data.append(np.concatenate([data for i, data in enumerate(data_list) if labels[i] in range(start, end + 1)]))
        grouped_labels.append(f"{start}-{end}")

    # Fig size setup
    fig, ax = plt.subplots(figsize=(15, 8))

    # Ploting
    VP = ax.boxplot(grouped_data, positions=np.arange(len(grouped_data)), widths=0.25,
                    patch_artist=True, showmeans=True, showfliers=True,
                    medianprops={"color": "white", "linewidth": 0.5},
                    boxprops={"facecolor": "C0", "edgecolor": "white",
                              "linewidth": 0.5},
                    whiskerprops={"color": "C0", "linewidth": 1.5},
                    capprops={"color": "C0", "linewidth": 1.5})

    # Scales and axis
    ax.set(xlabel='snp', ylabel='log(gene_expression)' if log_scale else 'gene_expression',
           xticks=np.arange(len(grouped_data)), xticklabels=grouped_labels,
           ylim=(0, y_upper_limit) if log_scale and y_upper_limit is not None else None,
           yticks=np.arange(0, 6) if log_scale else None,
           yticklabels=[f'$e^{{{i}}}$' for i in np.arange(0, 6)] if log_scale else None)
    
    # Set title
    if title:
        ax.set_title(title)
    
    plt.show()
    fig.savefig("{}_plot.pdf".format(title),dpi=300, bbox_inches="tight")
####################################################################################
```

```python
D=[df[df['snp'] == i]['gene_expression'].values for i in range(0,df.iloc[-1,0])]
plot_boxplot_grouped(D, labels=list(range(0,df.iloc[-1,0])),y_upper_limit=15,title="Taihu98:snp(interval)-gene_expression",
                     snp_ranges=[(0,0),(1,3),(4,6),(7,10),(11,15),(16,25),(25,30),(31,40),(41,50),(51,60)
                                 ,(61,80),(81,100),(101,150),(151,200),(201,250),(251,300)
                                 ,(301,350),(350,496),(0,496)])
```

```python
D=[df[df['snp'] == i]['gene_expression'].values for i in range(0,df.iloc[-1,0])]
plot_boxplot_grouped(D, labels=list(range(0,df.iloc[-1,0])),y_upper_limit=15,title="Taihu98:snp-gene_expression",
                     snp_ranges=[(0,0),(1,496),(0,496)])
```
